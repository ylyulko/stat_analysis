Consider the following pdf:<br>
$$
f(x)=
\begin{cases}
x\cdot e^{-x},\ x\geq 0,\\
0, \ x<0
\end{cases}
$$
Create a sample of size 1 000 from given  distribution and check via histogram that modelling is done correctly.

```python
def custom_pdf(x):
    res = 0
    if x > 0:
        res = x*np.exp(-x)
    return res
        
# print(custom_pdf(0.01))
```

draw the graph of pdf

```python
x_ = np.linspace(-2,10, 5000)
fx_ = [custom_pdf(xi) for xi in x_]


plot = go.Figure()
graph = go.Scatter(x = x_, y = fx_, name = 'Custom pdf')
plot.add_trace(graph)
```

Understand the logic which scipy.stats uses.

```python
class custom(rv_continuous):
    def _pdf(self, x):
        return custom_pdf(x)
```

Create object of the class and go!

```python
new_distr = custom()
```

does everything work? (check one by one)

```python
new_distr.pdf(1)
new_distr.cdf(1)
# new_distr.mean()
new_distr.median()
```

Generate sample of size 1000 and check that it corresponds to pdf specified

```python
sample = new_distr.rvs(size = 1000)
```

```python
x_ = np.linspace(-2,10, 5000)
fx_ = [custom_pdf(xi) for xi in x_]

plot = go.Figure()
graph = go.Scatter(x = x_, y = fx_, name = 'Custom pdf')
hist = go.Histogram(x = sample, histnorm = 'probability density')
plot.add_trace(hist)
plot.add_trace(graph)
```

**Bank account** class creation. Basic OOP principles.

```python
class bank_account:
    def __init__(self, r):
        self.rate = r
        
    def is_widthdrawable(self):
        pass  
```

```python
acc = bank_account(0.12)
acc.rate
acc.is_widthdrawable()
```

```python
acc2 = bank_account(0.2)
acc2.rate
```

```python
class deposit(bank_account):
    def __init__(self, r, a, T):
        super().__init__(r)
        self.amount = a
        self.maturity = T
        
    def is_widthdrawable(self):
        return True    
    
    def payment(self, t):
        return self.amount*np.exp(self.rate * t)
```

```python
dep = deposit(0.15, 1000, 1)
```

```python
dep_plus = deposit(T = 0.8, r = 0.12, a = 500)
dep_plus.rate
```

```python
dep.rate
dep.maturity
dep.is_widthdrawable()
dep.payment(0.5)
```
