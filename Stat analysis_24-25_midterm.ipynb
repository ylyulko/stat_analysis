**Question 1 [17 pts]**. Consider Markov chain $(C_n)_{n\ge 0}$ with $C_0=0$ states $S=\{0, 2\}$ and the following transitional probabilities: $p_{0,0}=1/8,\,p_{0,2}=7/8,\, p_{2,0}=3/4,\,p_{2,2}=1/4$. Find final probabilities $\pi_0,\,\pi_2$ numerically.</br>
Output: numerical values of $\widehat{\pi_0},\,\widehat{\pi_2}$ based on distribution of $C_{100}$ using $N=1000$ simulations. If you found theoretical values of $\pi_0,\,\pi_2$ in 2c) additionally output $|\pi_0-\widehat{\pi_0}|,\,|\pi_2-\widehat{\pi_2}|$.

```python
n = 100
N = 1000

ends = []
for _ in range(N):
    path = [0]
    for i in range(n+1):
        if path[i] == 0:
            new = random.choices([0, 2], weights = [0.125, 0.875])[0]
            path.append(new)
        elif path[i] == 2:
            new = random.choices([0, 2], weights = [0.75, 0.25])[0]
            path.append(new)
  
    ends.append(path[-1])

pi_1 = ends.count(0)/N
pi_2 = ends.count(2)/N
print(pi_1, pi_2)
print(abs(pi_1-6/13), abs(pi_2-7/13))
```

**Question 2 [16 pts]** Simulation of process $X_t=W_t^2-2t,\, t\ge 0,$ where $(W_t)_{t\ge 0}$ is Brownian motion.

```python
def bm_simulations(n_paths, granularity, max_time):
    n_steps = granularity * max_time
    
    paths = []
    for i in range(n_paths):
        xs = random.choices([-1, 1], weights = [0.5, 0.5], k = n_steps)
        xs = [0] + xs
        path = np.cumsum(xs)/np.sqrt(granularity)
        paths.append(path)
        
    return paths
```

1) Use function bm_simulations() to simulate $N=5$ paths of Brownian motion $W_t,\, 0\le t \le 1$. Use $granulatiry = 1000$. Modify obtained paths and get paths of $X_t,\, 0\le t \le 1.$ </br>
Output these paths in form of scatterplots.

```python
N = 5
n = 1000

t = np.linspace(0, 1, n+1)
paths = bm_simulations(N,n,1)
paths_x = [path**2 - 2*t for path in paths]
```

```python
plot = go.Figure()

for i in range(0,N):
    graph = go.Scatter(x = t, y = paths_x[i])
    plot.add_trace(graph)
    
plot.show()
```

Obtain $X_1= W_1^2-2$ and output sample mean of $X_1$. If you solved problem 3b) output difference $\left|sample\_mean(X_1) - E(X_1)\right|$

**Question 3 [17 pts]** Simulate stochastic differential equation
$$
dX_t = -\sqrt{t}X_tdt-3dW_t,\quad X_0=2.
$$

Using step $h=0.01$ simulate $N=2\,000$ trjectories of $X_t,\,t\in [0,1]$.</br>
Output sample mean of $X_1$. If you solved problem 4) output difference $|\mathsf E(X_1)-\overline{X_1}|$

```python
T = 1
n = 100
dt = 1/n

N = 2000

times = np.linspace(0, T, n*T+1)

paths = []
for _ in range(N):
    path = [2]
    for i in range(1, n*T + 1):
        ksi = norm.rvs(loc = 0, scale = np.sqrt(dt))
        ri = path[-1] - np.sqrt(times[i-1])*path[-1]*dt -3*ksi
        path.append(ri)
        
    paths.append(path)
```

```python
numbers = [path[-1] for path in paths]
s_mean = np.mean(numbers)
print(s_mean, abs(s_mean-2*np.exp(-2/3)))
```
