## Monte-Carlo methods

```python
import numpy as np
from scipy.stats import bernoulli
import time
```

Coin is sequentially fliped. Compute average number of coinflips until first HT

```python
# bernoulli.rvs(0.5)
```

```python
def experiment():
#     Let 1 be the H, 0 be the T.
    prev = bernoulli.rvs(0.5)
    X = 1
    stop = False
    while stop == False:
        current = bernoulli.rvs(0.5)
        X += 1
        if (prev == 1) & (current == 0):
            stop = True
        else:
            prev = current        
    return X
```

```python
start_time = time.time()

N = 100000
results = [experiment() for i in range(N)]
ans = np.mean(results)
print(ans)

# executed code to measure
print("--- %s seconds ---" % (time.time() - start_time))
```
