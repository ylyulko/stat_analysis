Write base class figure with methods *area(), perimeter(), center_distance(x,y), belong(x, y)*. What first two methods do is clear. The last method *center_distance(x,y)* calculates distance from center of the figure to point with coordinates (x,y).
Constructor of figure class takes two values (x0, y0) – center of figure. 

```python
class figure():
    def __init__(self, x0, y0):        
        self.center_x = x0
        self.center_y = y0
    
    def center_distance(self, x, y):
        return np.sqrt((x - self.center_x)**2 + (y - self.center_y)**2)
    
    def area(self):
        pass
    
    def perimeter(self):
        pass
    
    def belong(self,x,y):
        pass
```

```python
fig = figure(1,2)
# type(fig)
fig.center_y
fig.center_distance(0,0) 
# np.sqrt(5)
# fig.area()
```

Write child class *rectangular* and implement methods from base class.

```python
class rectangular(figure):
    def __init__(self, x0, y0, l, w):
        super().__init__(x0, y0)
        self.length = l
        self.width = w
    
#     override area() function of a base class
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2*(self.length+self.width)
    
    def belong(self,x,y):
        is_length = abs(x - self.center_x) < self.length/2
        is_width = abs(y - self.center_y) < self.width/2
        return is_length and is_width
```

Write child class square and implement methods from base class.

```python
rect = rectangular(0,0, 3, 5)
rect.center_x
rect.width
rect.center_distance(0, 3)
rect.belong(1,1)
rect.belong(1, 6)
rect.area()
rect.perimeter()
```

```python
class square(rectangular):
    def __init__(self, x0, y0, a):
        super().__init__(x0, y0, a, a)
```

```python
sq = square(0,0,3)
sq.area()
sq.perimeter()
```

Ellipse implementation

```python
class ellipse(figure):
    def __init__(self, x0, y0, a, b):
        super().__init__(x0, y0)
        self.r1 = a
        self.r2 = b
    def area(self):
        return np.pi * self.r1 * self.r2
    def perimeter(self):
        return 2 * np.pi * np.sqrt((self.r1**2 + self.r2**2)/2)
    def belong(self, x, y):
        return (x-self.center_x)**2/(self.r1**2) + (y-self.center_y)**2/(self.r2**2) <= 1
```

```python
ell = ellipse(0, 0, 3, 5)
ell.area()
ell.perimeter()
ell.belong(1, 1)
```

Circle implementation

```python
class circle(ellipse):
    def __init__(self, x0, y0, r):
        super().__init__(x0, y0, r, r)
```

```python
cir = circle(0, 0, 3)
cir.area()
cir.perimeter()
cir.belong(100, 100)
```
