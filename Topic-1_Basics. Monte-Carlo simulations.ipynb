**Graphs**

Construct a curve defined by 
$$
\begin{cases}
x = 2cos(t)+5cos(2t/3),\\
y = 2sin(t)-5sin(2t/3)
\end{cases}
$$

```python
t = np.linspace(-10, 10, 201)
# print(t)
x_ = 2*np.cos(t)+5*np.cos(2*t/3)
y_ = 2*np.sin(t)-5*np.sin(2*t/3)
```

**Random variables simulations, pdfs, cdfs**

In scipy.stats find functionality to generate Fisher distribution. Generate sequence of 10 000 random variables of $F(6,3)$ distribution.

```python
sample = f.rvs(6, 3, size = 10000)
sample
```

Construct histogram for generated sample and attach scatterplot of $F(6,3)$ pdf.

```python
plot = go.Figure()

hist = go.Histogram(x = sample, histnorm = 'probability density', name = 'histogram')
plot.add_trace(hist)

# Add pdf
x_ = np.linspace(0, 20, 1001)
y_ = f.pdf(x_, 6, 3)
fisher_scat = go.Scatter(x = x_, y = y_, name = 'pdf')
plot.add_trace(fisher_scat)

plot.update_xaxes(range = [0,20])
plot.show()
```

## Basic Monte-Carlo

**Integrals**

Compute integral 
$$
\int_1^5 x^3 ln^2(x) e^{-x}dx
$$

```python
def g(x):
    return (x**3)*(np.log(x)**2)*np.exp(-x)
```

Quickest way:

```python
N = 1000000
sample = uniform.rvs(loc = 1, scale = 4, size = N)
integral = (5-1)*np.mean(g(sample))
integralk
```

```python
x_ = np.linspace(1,5, 10001)
max(g(x_))

plot = go.Figure()
graph = go.Scatter(x = x_, y = g(x_))
plot.add_trace(graph)
plot.show()
```

```python
M = 2.5
a = 1
b = 5
N = 1000000
sample_x = uniform.rvs(loc = 1, scale = 4, size = N)
sample_y = uniform.rvs(loc = 0, scale = M, size = N)

R = sum(sample_y <= g(sample_x))
ans = (b-a)*R*M/N
ans
```
