### Random walk simulation

https://docs.python.org/3/library/random.html

```python
a = [0] + random.choices([-1, 1], weights = [0.6, 0.4], k = 10)
print(a)
print(np.cumsum(a))
```

```python
def rw_simulations(n_paths, max_time, p):
    paths = []    
    for _ in range(n_paths):
        xs = [0] + random.choices([-1, 1], weights = [1-p, p], k = max_time)
        path = np.cumsum(xs)
        paths.append(np.array(path))
        
    return np.array(paths)
```

```python
n = 50
p = 0.51 
N = 3
x_ = np.arange(n+1)
# x_

y_ = rw_simulations(N, n, p)
y_

plot = go.Figure()
for i in range(N):
    graph = go.Scatter(x = x_, y = y_[i])
    plot.add_trace(graph)
    
plot.show()
```

```python
def test_mean_rw(N, n, p, tolerance):
    
    theor_mean = (2*p-1)*n
    
    paths = rw_simulations(N, n, p)
    rn = paths[:, -1]
#   rn = [path[-1] for path in paths]
    is_mean_equal = abs(np.mean(rn) - theor_mean) < tolerance
    
    return is_mean_equal 
```

```python
n = 9
N = 1_000_000
p = 0.6

test_mean_rw(N, n, p, 0.01)
```

**Markov chains**

```python
C0 = 1
n = 8

path = [C0]
if C0 == 1:
    new_el = random.choices([1, 2], weights = [0.3, 0.7])[0]
if C0 == 2:
    new_el = 1
path.append(new_el)
```

```python
def chain(C0, max_time, n_paths):
    paths = []
    for _ in range(n_paths):
        path = [C0]
        for i in range(max_time):
            if path[-1] == 1:
                new_el = random.choices([1, 2], weights = [0.3, 0.7])[0]
            if path[-1] == 2:
                new_el = 1
            path.append(new_el)
        paths.append(np.array(path))
        
    return np.array(paths)
```

```python
C0 = 1
n = 120
N = 10_000

paths = chain(C0, n, N)
pi1 = sum(paths[:, -1] == 1)/N
pi2 = sum(paths[:, -1] == 2)/N

print(pi1, pi2, abs(pi1-10/17), abs(pi2-7/17))
```
