**Question 1 [15 pts]**. Compute expectation $\mathsf E(2^X)$ where $X\sim Geom(1/4)$ via Law of Large numbers (or Monte Carlo method)</br>
$$
P(X=k)=\left(\displaystyle\frac{3}{4}\right)^{k-1}\cdot\displaystyle\frac{1}{4},\, k\ge 1
$$
Use $n=10\,000$ simulations, output computed approximation of expectation.

```python
def experiment():
    val = geom.rvs(p = 1/4)
    return 2**val

values = []
for i in range(0,10_000):
    values.append(experiment())
    
np.mean(values)
```

**Question 2 [22 pts]** Simulation of modulus $|W_t|,\, t\ge 0$ of Brownian motion.

```python
def bm_simulations(n_paths, granularity, max_time):
    n_steps = granularity*max_time

    paths = []
    for _ in range(0, n_paths):
        seq = np.array(random.choices([-1, 1], weights = [0.5, 0.5], k = n_steps))
        path = np.cumsum(seq)/np.sqrt(granularity)
        # insert 0 to the beginning as an entry point
        path = np.insert(path,0,0)
        paths.append(path)    
   
    return paths
```

1) Use function bm_simulations() to simulate $N=10$ paths of Brownian motion $W_t,\, 0\le t \le 1$. Use $granulatiry = 1000$. Modify obtained paths and get paths of modulus $|W_t|,\, 0\le t \le 1.$ </br>
Output these paths in form of scatterplots.

```python
N = 10
n = 1000

paths = bm_simulations(N,n,1)
paths_modulus = np.abs(paths)
```

```python
t = np.linspace(0, 1, n+1)

plot = go.Figure()

for i in range(0,N):
    graph = go.Scatter(x = t, y = paths_modulus[i])
    plot.add_trace(graph)
    
plot.show()
```

**Question 3 [15 pts]** Simulate stochastic differential equation
$$
dX_t = e^{-t/2}dt+4dW_t,\quad X_0=2.
$$

Using step $h=0.01$ simulate $N=10\,000$ trjectories of $X_t,\,t\ge 0$.</br>
Output sample mean of $X_1$. If you solved problem 4) output theoretical mean $\mathsf E(X_1)$

```python
from scipy.stats import norm
```

```python
N = 10000
m = 100
h = 1/m

paths = []

for _ in range(N): 
    deltas = norm.rvs(loc = 0, scale = np.sqrt(h), size = m)
    path = [2]
    for i in range(1, m+1):
        si = path[i-1] + np.exp(-0.5*(i-1)*h)*h + 4*deltas[i-1]
        path.append(si)
    paths.append(path)
```

```python
numbers = [path[-1] for path in paths]
print(np.mean(numbers), 4-2*np.exp(-0.5))
```
